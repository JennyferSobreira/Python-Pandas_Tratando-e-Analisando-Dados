## Seleções e Frequências

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
```

```python
#Selecione somente os imóveis classificados com tipo 'Apartamento'.
selecao = dados['Tipo'] == 'Apartamento'
n1 = dados[selecao]
n1
```

```python
n1_shape = dados[selecao].shape[0]
n1_shape
```

```python
#Selecione os imóveis classificados com tipos 'Casa', 'Casa de Condomínio' e 'Casa de Vila'.
selecao = (dados['Tipo'] == 'Casa') | (dados['Tipo'] == 'Casa de Condomínio') | (dados['Tipo'] == 'Casa de Vila')
n2 = dados[selecao]
n2
```

```python
n2_shape =  dados[selecao].shape[0]
n2_shape
```

```python
#Selecione os imóveis com área entre 60 e 100 metros quadrados, incluindo os limites.
# 60 <= Area <= 100
selecao = (dados['Area'] >= 60) & (dados['Area'] <= 100)
n3 = dados [selecao]
n3
```

```python
n3_shape = dados [selecao].shape[0]
n3_shape
```

```python
#Selecione os imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00.
seleca = (dados['Quartos'] >= 4) & (dados['Valor']< 2000)
n4 = dados[selecao]
n4
```

```python
n4_shape = dados[selecao].shape[0]
n4_shape
```

```python
print("Número de imóveis  classificados com tipo 'Apartamento'. -> {}.".format(n1_shape))

print("Número de imóveis classificados com tipos 'Casa', 'Casa de Condomínio' e 'Casa de Vila'. -> {}.".format(n2_shape))

print("Número de imóveis com área entre 60 e 100 metros quadrados, incluindo os limites. -> {}.".format(n3_shape))

print("Número de imóveis  que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00. -> {}.".format(n4_shape))
```
